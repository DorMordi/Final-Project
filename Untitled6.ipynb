```python
# The function get locations list and limit of number of restaurants 
# The function call to Api_Request function and merge the dataframes
# The function return merged dataframe with all the data
def Get_DataFrame(locations, limit):
    frames = []
    for location in locations:
        for i in range (0, limit, 50):
            frames.append(Api_Request(location,i))
    df = pd.concat(frames, ignore_index=True)
    return df
```

```python
# The fucnction get dataframe that in column 'categories' have a list of up to three diffrent categories
# The function take the first category of each restaurant assigment it in new cateogty columns
def Categories_Order(df):
    category = []
    for cat in df['categories']:
            category.append(cat[0]['alias'])
    df['category'] = category
    del df['categories']
    return df
```

```python
# The function transfer to numeric the price of the restaurant ($, $$, $$$, $$$)
def Price_Order(df):
    replace_map = {'$':1, '$$':2, '$$$':3, '$$$$':4}
    df.replace(replace_map, inplace = True)
    df['price'] = df['price'].fillna(df['price'].mode()[0])
    return df
```

```python
# The function splitting the location of the restaurant to city, state and zip code columns
def Location_Order(df):
    city = []
    state = []
    #country = []
    zip_code = []
    for loc in df['location']:
        city.append(loc['city'])
        state.append(loc['state'])
    #    country.append(loc['country'])
        zip_code.append(loc['zip_code'])
    df['city'] = city
    df['state'] = state
    #df['country'] = country
    df['zip_code'] = zip_code
    del df['location']
    return df
```

```python
# The function splitting the transaction column to delivery, pickup and restaurant reservation columns
def Tansaction_Order(df):
    delivery = []
    pickup = []
    restaurant_reservation = []
    for tran in df['transactions']:
        if 'delivery' in tran:
            delivery.append(1)
        else:
            delivery.append(0)
        if 'pickup' in tran:
            pickup.append(1)
        else:
            pickup.append(0)
        if 'restaurant_reservation'in tran:
            restaurant_reservation.append(1)
        else:
            restaurant_reservation.append(0)
    df['delivery'] = delivery
    df['pickup'] = pickup
    df['restaurant reservation'] = restaurant_reservation
    del df['transactions']
    return df
```

```python
# The funtion get categories list and remove business from df which inclusd in the list
def Remove_Unessery_Rows_By_Category(df, categories):
    for category in categories:   
        df = df[df.category != category]
    df = df.reset_index(drop = True)
    return df
```

```python
# The function remove restaurants which not found in the locations list: New York, Miami, Los Angeles
def Remove_unnecessey_Rows_By_City(df, locations):
    df = df.drop(df.index[~df['city'].isin(locations)])
    df = df.reset_index(drop = True)
    return df
    
```

```python
# The function get list of unnecessey columns and remove them from the df
def Delete_Unused_Parameters(df, columns):
    for col in columns:
        del df[col]
    return df
```

```python
# The function get column and transfer the str values to numeric values
def Transfer_Str_To_Numeric_Values(df, column):
    c = df[column].astype('category')
    dictionary = dict(enumerate(c.cat.categories))
    df[column] = df[column].astype("category").cat.codes
    return df, dictionary
```

```python
# The function get list of columns and normalize them by MinMaxScaler
def Normalize_Columns(df, columns):
    for col in columns:
        df[col] = MinMaxScaler().fit_transform(np.array(df[col]).reshape(-1,1))
    return df
```

```python
# The function split the dataframe to X and y 
def X_y_Split(df, label_column):
    TRAINING_FEATURES = df.columns[df.columns != label_column]
    X = df[TRAINING_FEATURES]
    y = df[label_column]
    return X, y
```

```python
# The function using train_test_split to X_Train, X_test, y_train, y_test
def Split_To_Train_And_Test(X, y, test_ratio, rand_state):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_ratio, random_state = rand_state)
    return X_train, X_test, y_train, y_test
```

```python
# The function create stnadard scaler and return X_train_scaled
def Standard_Scaler(X_train):
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    return scaler, X_train_scaled
```

```python
# The function build Linear Regression model
def Train_Classifier(X_train, y_train):
    clf_model = LinearRegression().fit(X_train, y_train)
    return clf_model
```

```python
# The function predict y_pred by X_test
# The function return dataframe if actual results (y_yest) and predicted results (y_pred)
def Predict(classifier, X_test, y_test):
    y_pred = classifier.predict(X_test)
    resDF  = pd.DataFrame({"Actual":(y_test)/2,"Predicted":(y_pred)/2})
    return resDF, y_pred
```
